#### Question 4: Write a Python function that returns the items having highest % (daily value) of Vitamin A

```python
def get_highest_vitamin_A(data):
    max_vitA = max(data['Vitamin A (% Daily Value)']) #Get the maximum % daily value of Vitamin A
    cond = (data['Vitamin A (% Daily Value)'] == max_vitA) 
    rows = data.loc[cond] #Select all rows satisfying the specified condition
    items = [ x[1] for x in rows.values if x[20] == max_vitA]  
    return items
```

#### Question 5: Write a Python function that returns the average amount of 'Sugars' in each 'Category'. The output should be a dictionary, whose keys are categories and values are the average amount of sugar in that category.

```python
def get_average_sugars(data):
    categories = data.groupby('Category')
    category_means = categories['Sugars'].mean() #Find the mean sugars for each category
    return category_means.to_dict() #Convert to dictionary
```

```python
average_sugars = get_average_sugars(df)
average_sugars
```
